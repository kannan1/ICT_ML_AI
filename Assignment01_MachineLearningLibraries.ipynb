#Data Visualization

```python
import matplotlib.pyplot as plt
```

#Univariate Analysis

```python
plt.figure(figsize=(5,3))
plt.boxplot(df['Age'])
plt.ylabel('OutCome')
plt.xlabel('Age')
plt.show()
```

```python
import seaborn as sns
```

#Bivariate Analysis

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x='Age',y='BMI', data= df)
plt.xlabel('Age')
plt.ylabel('BMI')
plt.title('Age vs BMI')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Glucose'])
plt.xlabel('Glucose')
plt.ylabel('Frequency')
plt.show()
```
